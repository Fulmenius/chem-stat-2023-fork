## Часть 3. Подготовка таблицы для работы.

Давайте теперь подготовим таблицу, чтобы нам было с ней удобнее работать. Как мы видим, в первой колонке таблицы выше приведены номера лабораторий, в которых проводились измерения. В следующих трех колонках приведены значения содержаний этих элементов. Выделим именно этот кусок из исходной таблицы, чтобы работать только с ним. 

Заведём новый датафрейм `dfChanged`. Подробнее про метод `iloc` смотрите [тут](https://sparkbyexamples.com/pandas/pandas-iloc-usage-with-examples/).

```python
dfChanged = df.iloc[0:61, 0:4]
```

```python
dfChanged.head()
```

Теперь давайте заведём новые заголовки талицы, которые являются тем, что написано в строке с индексом `0`.

### Средневзвешенное

$$ \overline{x} = \dfrac{\sum_{i = 1}^N w_i x_i}{\sum_{i = 1}^N w_i}$$

Давайте создадим какой-нибудь список весов. Узнайте, как работает метод `range()`

```python
w = []
for item in range(60):
    w.append((item + 1)/60)
```

```python
print(w)
```

```python
len(w)
```

Найдём среднее взвешенное значение, используя метод `np.average()`, в качестве аргументов метода передадим столбец `al` и веса `w`.

```python
np.average(al, weights= w)
```

Иногда оказывается полезным посмотреть на взаимосвязь нескольких признаков. Как нарисовать и гистограммы, и узнать, есть ли взаимосвязь между признаками?

Давайте вычислим процентили. Процентили -- это значение, которое заданная случайная величина не превышает с фиксированной вероятностью.

```python
#Обратите внимание, что мы используем флаг pct = True

dfChanged['percentile'] = dfChanged['Al'].rank(pct = True)
```

```python
dfChanged.head()
```

```python
#Заведём новый датафрейм, в котором будут только значения аллюминия и процентили

dfControl = dfChanged[['Al', 'percentile']]
```

```python
dfControl.head()
```

Теперь нам надо отсортировать таблицу по возрастанию  

```python
dfControl = dfControl.sort_values(by = 'percentile', ascending=True)
```

```python
dfControl.head()
```

```python
dfControl.tail()
```

Нам нужно построить нормальную бумагу. Для этого нам надо взять значния из столбца процентилей, и вызывать функцию, обратную к функции распределения.

```python
dfControl['res'] = scipy.stats.norm.ppf(dfControl['percentile'])
```

```python
dfControl.head()
```

А теперь давайте построим точечный график того, что получается.

```python
plt.pyplot.plot(dfControl['Al'], dfControl['res'], 'o')
```

И наконец, построим непрерывный график. 

```python
plt.pyplot.plot(dfControl['Al'], dfControl['res'])
```

Что мы видим? Что укладка графика почти что линейная. А значит, мы можем сделать вывод о нормальности распределения!
